# Часть вторая

**Проводим измерения для большого количества точек**

Расчеты проводились с временем симуляции T = 0.0001

Шаг dt расчитвался как piece_width^2 / 2

,где piece_width = 1 / pieces_amount

```python
for pieces_amount in time_data["pieces_amounts"].unique():
    pieces_data = time_data[time_data["pieces_amounts"] == pieces_amount]
    plt.figure(figsize=(15,10))
    plt.plot(pieces_data.proccesses,
             pieces_data.duration)
    plt.title("Test " + str(pieces_amount) + " pieces amount duration")
    plt.xlabel("Proccesses")
    plt.ylabel("duration (seconds)")
    plt.grid()
    plt.show()
```
