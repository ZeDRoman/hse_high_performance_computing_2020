```python
import pandas as pd
import matplotlib.pyplot as plt

time_data = pd.read_csv("data/time_data.out", names=['threads_amount', 'task', 'duration'])
```

```python
for i in range(1, 6):
    points_data = pd.read_csv("data/data" + str(i) + ".out", names=['x', 'y', 'cluster'])
    plt.figure(figsize=(10,10))
    plt.title("Test: " + str(i) + " clusters")
    plt.scatter(points_data.x, points_data.y, c=points_data.cluster, cmap="prism")
    plt.show()
    
    plt.figure(figsize=(10,10))
    test_time_data = time_data[time_data.task == i]
    plt.plot(test_time_data.threads_amount, test_time_data.duration)
    plt.title("Test " + str(i) + " duration")
    plt.xlabel("Threads")
    plt.ylabel("duration (microseconds)")
    plt.grid()
    plt.show()
```

```python
plt.figure(figsize=(10,5))
speed_up_data = time_data.groupby("threads_amount").sum()
speed_up_data.duration = speed_up_data.loc[[1]]["duration"].values[0] / speed_up_data.duration 
plt.plot(speed_up_data.index, speed_up_data.duration)
plt.scatter(speed_up_data.index, speed_up_data.duration)
plt.title("Summary tests time speedup")
plt.xlabel("Threads")
plt.ylabel("speedup")
plt.grid()
plt.show()
```
